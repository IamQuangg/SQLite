```python
# Chon du lieu tran dau bao gom thong tin giai dau va quoc gia
pd.read_sql_query(''' Select * from Match m 
                      Join league as l on m.league_id = l.id
                      Join country as c on c.id =m.country_id''',conn)
```

```python
# Lay so luong tran dau cua moi giai dau, bao gom ca ten giai dau, sap xep theo thu tu giam dan so tran dau
pd.read_sql_query('''Select l.name
                    ,count(*)  as game_play
                    from Match m 
                    Join league as l on m.league_id = l.id
                    Join country as c on c.id =m.country_id
                    Group by l.name
                    Order by game_play DESC
                      ''',conn)
```

```python
# lay tong ban thang cua doi nha va doi khach trong moi giai dau
pd.read_sql_query('''Select l.name
                    ,Sum(m.home_team_goal)  home_goal
                    ,Sum(m.away_team_goal) away_home
                    from Match m 
                    Join league as l on m.league_id = l.id
                    Join country as c on c.id =m.country_id
                    Group by l.name
                      ''',conn)
```

```python
# Chon du lieu tu bang Team
pd.read_sql_query('''Select * from team ''',conn)
```

```python
# Loc ra top 20 doi co so ban thang tren san nha nhieu nhat
pd.read_sql_query('''Select t.team_long_name
                    ,Sum(m.home_team_goal) home_goal
                    from Match m 
                    Join team t on t.team_api_id = m.home_team_api_id
                    Group By t.team_long_name
                    Order By Sum(m.home_team_goal) DESC
                    Limit 20 
                     ''', conn)
```

```python
# Loc ra top 20 doi co so ban thang tren san khach cao nhat
pd.read_sql_query('''Select t.team_long_name
                    ,Sum(m.away_team_goal) away_goal
                    from Match m 
                    Join team t on t.team_api_id = m.away_team_api_id
                    Group By t.team_long_name
                    Order By Sum(m.away_team_goal) DESC
                    Limit 20 
                     ''', conn)
```

```python
# Liet ke cac doi bong va tong so ban thang cua moi doi, sap xep theo so luong giam dan
pd.read_sql_query(''' Select team_long_name
                    , sum(goal) total_goal from 
                    (
                    Select team_long_name
                    ,Sum(m.home_team_goal) goal
                    From team t 
                    Join match m 
                    on t.team_api_id = m.home_team_api_id 
                    Group By team_long_name
                     Union ALL
                     Select team_long_name
                    ,Sum(m.away_team_goal) goal
                    From team t 
                    Join match m 
                    on t.team_api_id = m.away_team_api_id 
                    Group By team_long_name
                    )
                    Group By team_long_name
                    Order by total_goal DESC
                       ''',conn)
```

```python
# Liet ke ten cac doi bong va tong so tran dau 
pd.read_sql_query(''' Select team_long_name
                     ,Sum(game_play) total_play
                     From (
                    Select t.team_long_name
                    ,Count(m.match_api_id) game_play
                    from match m 
                    join team t
                    On m.home_team_api_id = t.team_api_id
                    Group By t.team_long_name 
                    Union ALL
                     Select t.team_long_name
                    ,Count(m.match_api_id) game_play
                    from match m 
                    join team t
                    On m.away_team_api_id = t.team_api_id
                    Group By t.team_long_name)
                    Group By team_long_name
                    Order By total_play DESC  ''',conn)
```

```python
# Liet ke tong so ban thang ghi duoc cua tung giai dau theo tung mua giai 
pd.read_sql_query('''Select name
                    ,season
                    ,sum(home_team_goal + away_team_goal) total_goal
                    from match m 
                    join league l 
                    on m.league_id = l.id 
                    Group By name,season ''',conn)
```
